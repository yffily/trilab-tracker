```python
def get_frame_info(input_file):
    print(input_file)
    cap     = cv2.VideoCapture(input_file)
    Nframes = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    i = 0
    while cap.grab():
        i += 1
    cap.release()
    return { 'File':os.path.basename(input_file), 
             'Total frames':Nframes,
             'Readable frames': i,
             'Mismatch': i!=Nframes }


with multiprocessing.Pool(None) as pool:
    frame_info = pool.map(get_frame_info,input_files)


writer = pd.ExcelWriter('frame-info.xlsx')
pd.DataFrame(frame_info).to_excel(writer,sheet_name='sheet1',index=None)
sheet = writer.sheets['sheet1']
for col,width in zip('ABC',[35,25,25]):
    sheet.column_dimensions[col].width = width
writer.save()
```
